# Visit Counting: Approach 1

Using whole data set



1.   Left spatial joint between the whole data set (left) and park data set (right). This returns a dataframe with an indicator of whether a row is inside the park or not.
2.   Sort dataframe by date time
3.   Add label column, which denotes whether a data point is inside the park or not.
4.   Group by each identifier and date, then add a secondary label which shows whether the data point has just moved in or out of the park.
5.   Delete all data points that are outside the park (i.e. keep those inside the park), then do a cumulative sum during groupby to get the number of visits per day.

Pros:
* More true to the data, so a visit is clearly defined (as a streak of data points that falls within the park's coordinate within a day, until a data point is seen outside of the park).

Cons:
* Require full data set.


```
# outer spatial joint with park shape file
big_df = gpd.sjoin(gdf, park, how="left", op='intersects')
big_df['date'] = big_df.local_date_time.apply(lambda x: x[:10])
```

```
# sort by date
big_df =  big_df.sort_values('local_date_time')
```

```
# count visits
big_df['label1'] = big_df.groupby(['identifier', 'date']).label.shift() # just the label, but shifted by 1
big_df.label = big_df.label != big_df.label1 # main comparison --> whether the label changes from the previous point 
big_df1 = big_df.dropna(subset=['index_right']) # drop all points that are outside of the park --> help with counting
big_df1['visit_count'] = big_df1.groupby(['identifier', 'date']).label.cumsum() # counting
```

# Visit Counting: Approach 2

Using distributions



1.   From the inner spatial joint between all data points and park data, group by identifier and date, then calculate the time difference between each data point and its previous record.
2.   Drop NaNs and select a quantile for visit cut-off (in this case, 75%).
3.   Any two data points from the same identifier at the same date but are further than the cut-off are denoted as two different visits.

Pros:
* Quick and doesn't require the complete set of data.

Cons:
* Visits not strictly defined and subject to the selection of parameter.

```
import datetime

# process df to get time_diff
to_plot['date'] = to_plot.local_date_time.apply(lambda x: x[:10])
to_plot =  to_plot.sort_values('local_date_time')
to_plot['time'] = to_plot.local_date_time.apply(lambda x: x[11:])
to_plot.time = to_plot.time.apply(lambda x: datetime.datetime.strptime(x, "%H:%M:%S"))
to_plot['next_time'] = to_plot.groupby(['identifier', 'date']).time.shift(-1)
to_plot['time_diff'] = to_plot.next_time - to_plot.time
to_plot.duration = to_plot.duration.apply(pd.to_timedelta, unit='s')
to_plot.time_diff = to_plot.time_diff.fillna(to_plot.duration)

to_plot.time_diff.apply(lambda x: x.total_seconds()).dropna().hist(bins=50)
```

# Trajectory Simplification



*   Trajectories seem simple enough
*   One idea to further simplify the trajectories is:


1.   If two successive data points from the same identifier from the same day are not fundamentally different, delete one of them
2.   "Fundamentally different" is defined to be X seconds away from each other, OR Y geometrical units away from each other. This way, an identifier standing at the same spot all day would still be recorded (since this info is more or less meaningful); similarly, a constantly moving identifier would be recorded properly with fewer data points.
